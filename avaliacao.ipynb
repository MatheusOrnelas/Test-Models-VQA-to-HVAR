# Previous analysis of pre-trained models

```python
import gradio as gr
from transformers import ViltProcessor, ViltForQuestionAnswering, BlipProcessor, BlipForConditionalGeneration
import torch

# Carregar os modelos e processadores padrão
vilt_processor = ViltProcessor.from_pretrained("dandelin/vilt-b32-mlm")
vilt_model = ViltForQuestionAnswering.from_pretrained("dandelin/vilt-b32-mlm")

blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

device = "cuda" if torch.cuda.is_available() else "cpu"
vilt_model.to(device)
blip_model.to(device)

def predict_vilt(image, question):
    """
    Predicts the answer to a question about an image using the ViLT model.

    Args:
        image (Pil.Image.Image): The image to analyze.
        question (str): The question about the image.

    Returns:
        The predicted answer to the question.
    """

    inputs = vilt_processor(images=[image], text=[question], return_tensors="pt", padding=True, truncation=True).to(device)
    outputs = vilt_model(**inputs)
    answers = outputs.logits.argmax(-1)
    answer = vilt_processor.tokenizer.decode(answers[0], skip_special_tokens=True)

    return answer

def predict_blip(image, question):
    """
    Generates a caption for an image using the BLIP model, conditioned on the provided question.

    Args:
        image (Pil.Image.Image): The image to analyze.
        question (str): The question to guide the caption generation.

    Returns:
        The generated caption for the image.
    """

    inputs = blip_processor(images=image, text=question, return_tensors="pt", padding=True, truncation=True).to(device)
    outputs = blip_model.generate(**inputs)
    prediction = blip_processor.tokenizer.decode(outputs[0], skip_special_tokens=True)

    return prediction

def predict_both_models(image, question):
    """
    Uses both ViLT and BLIP models to process an image and question, returning predictions from both.

    Args:
        image (Pil.Image.Image): The image to be analyzed.
        question (str): The question about the image.

    Returns:
        A tuple containing the predictions from the ViLT and BLIP models.
    """
    vilt_result = predict_vilt(image, question)
    blip_result = predict_blip(image, question)

    return vilt_result, blip_result


# Config the Gradio interface
iface = gr.Interface(
    fn=predict_both_models,
    inputs=[gr.Image(type="pil"), gr.Textbox(label="Question")],
    outputs=[gr.Textbox(label="ViLT Prediction"), gr.Textbox(label="BLIP Prediction")]
)

iface.launch(debug=True)
```

# Train Blip e ViLT

```python
import os
import torch
from transformers import ViltProcessor, ViltForQuestionAnswering, BlipProcessor, BlipForQuestionAnswering, TrainingArguments, Trainer
from datasets import load_dataset
from PIL import Image
from tqdm import tqdm
from transformers import DefaultDataCollator
from huggingface_hub import HfApi, Repository
import datasets
import transformers
import json


class FashionModelTrainer:
    def __init__(self, model_type="vilt", num_samples=110):
        """
        Initializes the FashionModelTrainer with a specified model type and sample size.

        Args:
            model_type (str): The type of model to use ('vilt' or 'blip'). Default is 'vilt'.
            num_samples (int): Number of samples to use from the dataset. Default is 110.
        """

        self.model_type = model_type
        self.num_samples = num_samples
        self.setup()

    def setup(self):
        """
        Sets up the model and data processor based on the selected model type and loads the dataset.
        """

        if self.model_type == "vilt":
            self.model_checkpoint = "dandelin/vilt-b32-mlm"
            self.processor = ViltProcessor.from_pretrained(self.model_checkpoint)
            self.model = ViltForQuestionAnswering.from_pretrained(self.model_checkpoint)

        elif self.model_type == "blip":
            self.model_checkpoint = "Salesforce/blip-vqa-base"
            self.processor = BlipProcessor.from_pretrained(self.model_checkpoint)
            self.model = BlipForQuestionAnswering.from_pretrained(self.model_checkpoint)
            self.model.to(torch.device("cuda" if torch.cuda.is_available() else "cpu"))

        self.dataset = load_dataset("SaffalPoosh/deepFashion-with-masks", split=f"train[:{self.num_samples}]")
        self.dataset = self.dataset.map(FashionModelTrainer.replace_underscores)
        self.data_collator = DefaultDataCollator(return_tensors="pt")


    def preprocess_data(self, dataframe):
        """
        Processes the data to be suitable for model training, including image and question preparation.

        Args:
            dataframe (dataset): A batch from the dataset to be processed.

        Returns:
            Processed data ready for training.
        """

        images = dataframe['images']
        total_questions = [
            "What is the person wearing?",
            "What kind of clothing is this?",
            "Can you identify the style of clothing?",
            "What category of clothing does this belong to?",
            "What’s the classification of this attire?",
            "What sort of clothing item is this?",
            "Could you specify the type of garment?",
            "What fashion type does this represent?",
            "In which clothing category would you place this?",
            "What is the designation for this style of clothing?",
            "What is the nature of this clothing piece?"
        ]

        # Distribute questions evenly across the images
        questions_per_image = len(images) // len(total_questions)
        questions = []
        for question in total_questions:
            questions.extend([question] * questions_per_image)

        # If there are any remaining images, add the last question to them
        remaining = len(images) % len(total_questions)
        if remaining:
            questions.extend([total_questions[-1]] * remaining)

        if self.model_type == "vilt":
            texts = dataframe['question']
            labels = [self.label2id[label] for label in dataframe['label']]
            encoding = self.processor(images, texts, padding="max_length", truncation=True, return_tensors="pt")
            one_hot_labels = torch.zeros(len(labels), len(self.label2id))
            for i, label in enumerate(labels):
                one_hot_labels[i, label] = 1
            encoding["labels"] = one_hot_labels

        elif self.model_type == "blip":
            captions = dataframe['caption']
            encoding = self.processor(images, questions, padding="max_length", truncation=True, return_tensors="pt", max_length=128)
            labels = self.processor.tokenizer(captions, padding="max_length", truncation=True, max_length=128, return_tensors="pt")
            encoding = {k: v.squeeze() for k, v in encoding.items()}
            encoding['labels'] = labels['input_ids'].squeeze()

        return encoding

    def train(self):
        """
        Configures the training settings and starts training the model.
        """

        if self.model_type == "vilt":
            self.prepare_vilt_dataset()
            self.training_args = TrainingArguments(
                output_dir="./vilt_finetuned_fashion",
                per_device_train_batch_size=4,
                num_train_epochs=20,
                # num_train_epochs=2,
                save_steps=10,
                logging_steps=10,
                learning_rate=5e-5,
                push_to_hub=True,
                save_total_limit=2,
            )
        elif self.model_type == "blip":
            self.prepare_blip_dataset()
            self.training_args = TrainingArguments(
                output_dir="./blip_finetuned_fashion",
                per_device_train_batch_size=4,
                num_train_epochs=10,
                #num_train_epochs=2,
                logging_steps=10,
                save_steps=10,
                evaluation_strategy="epoch",
                push_to_hub=True,
                save_total_limit=2,
            )

        self.trainer = Trainer(
            model=self.model,
            args= self.training_args,
            data_collator=self.data_collator,
            train_dataset=self.train_dataset,
            eval_dataset=getattr(self, 'valid_dataset', None),
            tokenizer=self.processor,
        )
        self.trainer.train()

        try:
          # It is not recommended to perform the push this way,
          # but for the sake of example, we will do the push like this.

          self.trainer.push_to_hub()
        except:
          print('Error when pushing to hub')


    def prepare_vilt_dataset(self):
        """
        Prepares the dataset for training with the VILT model.
        """

        self.dataset = self.dataset.rename_column("caption", "question")
        self.dataset = self.dataset.rename_column("cloth_type", "label")
        unique_labels = set(self.dataset['label'])
        self.label2id = {label: idx for idx, label in enumerate(unique_labels)}
        self.id2label = {idx: label for label, idx in self.label2id.items()}
        self.model = ViltForQuestionAnswering.from_pretrained(
            self.model_checkpoint,
            num_labels=len(unique_labels),
            id2label=self.id2label,
            label2id=self.label2id
        )
        self.train_dataset = self.dataset.map(self.preprocess_data, batched=True, remove_columns=self.dataset.column_names)


    def prepare_blip_dataset(self):
        """
        Prepares the dataset for training with the BLIP model.
        """

        self.train_dataset = self.dataset.map(self.preprocess_data, batched=True, remove_columns=self.dataset.column_names)
        self.valid_dataset = load_dataset("SaffalPoosh/deepFashion-with-masks", split=f"train[{self.num_samples}:{2 * self.num_samples}]").map(self.preprocess_data, batched=True, remove_columns=self.dataset.column_names)


    def replace_underscores(dataframe):
        """
        Replaces underscores in text fields within with spaces.

        Args:
            dataframe (dataset): Data from the dataset.

        Returns:
            The modified dataframe with underscores replaced by spaces in string fields.
        """

        for column in dataframe:
            if isinstance(dataframe[column], str):
                dataframe[column] = dataframe[column].replace("_", " ")

        return dataframe

```

# Evaluation of models after fine-tuning

```python
import torch
import gradio as gr
from transformers import AutoProcessor, AutoModelForVisualQuestionAnswering, BlipForQuestionAnswering

vilt_processor = AutoProcessor.from_pretrained("Ornelas/vilt_finetuned_fashion")
vilt_model = AutoModelForVisualQuestionAnswering.from_pretrained("Ornelas/vilt_finetuned_fashion")
blip_processor = AutoProcessor.from_pretrained("Ornelas/blip_finetuned_fashion")
blip_model = BlipForQuestionAnswering.from_pretrained("Ornelas/blip_finetuned_fashion")

vilt_model.eval()
blip_model.eval()

def predict_both_models(image, question):
    """
    Predicts answers from both ViLT and BLIP models for a given image and question.

    Args:
        image (Pil.Image.Image): The image to analyze.
        question (str): The question to be answered about the image.

    Returns:
        A tuple containing predictions from both ViLT and BLIP models.
    """
    # Device Config
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    vilt_model.to(device)
    blip_model.to(device)

    # Prediction with ViLT
    vilt_inputs = vilt_processor(images=[image], text=question, return_tensors="pt")
    vilt_inputs = {k: v.to(device) for k, v in vilt_inputs.items()}
    with torch.no_grad():
        vilt_outputs = vilt_model(**vilt_inputs)
    vilt_logits = vilt_outputs.logits
    vilt_probs = torch.nn.functional.softmax(vilt_logits, dim=-1)
    vilt_top_probs, vilt_top_labels = torch.topk(vilt_probs, 5)
    vilt_top_probs = vilt_top_probs[0].cpu().numpy().tolist()
    vilt_top_labels = [vilt_model.config.id2label[label_id] for label_id in vilt_top_labels[0].cpu().numpy().tolist()]
    vilt_results = "\n".join([f"{label}: {prob:.2%}" for label, prob in zip(vilt_top_labels, vilt_top_probs)])

    # Prediction with BLIP
    blip_inputs = blip_processor(images=[image], text=question, return_tensors="pt")
    blip_inputs = {k: v.to(device) for k, v in blip_inputs.items()}
    with torch.no_grad():
        blip_generated_ids = blip_model.generate(**blip_inputs)
    blip_decoded_preds = blip_processor.batch_decode(blip_generated_ids, skip_special_tokens=True)

    return vilt_results, blip_decoded_preds[0]

# Config Gradio interface
iface = gr.Interface(
    fn=predict_both_models,
    inputs=[gr.Image(type="pil"), gr.Textbox(label="Question")],
    outputs=[gr.Textbox(label="ViLT Predictions"), gr.Textbox(label="BLIP Predictions")]
)

iface.launch(debug=True)
```
